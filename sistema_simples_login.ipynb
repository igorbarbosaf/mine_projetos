```python
usuarios = {}

def menu():

    print("\n1 - Cadastrar usuário")
    print("2 - Fazer login")
    print("3 - Lista de usuários")
    print("4 - Sair")

def cadastrar_usuario():

    login = input("Digite o login: ").strip()

    if not login:
        print("Login não pode ficar sem preencher.")
        return

    if login not in usuarios:
        print("Cadastrando login com sucesso.")
        return
    
    senha = input("Digite sua senha: ")

    if not senha:
        print("Por favor, a senha não pode ficar vazia.")
        
    else:
        usuarios[login] = senha
        print("Login e senha cadastrando com sucesso.")

def fazer_login():

    login = input("Digite seu login: ").strip()
    senha = input("Digite sua senha: ")

    if login in usuarios:
        if usuarios[login] == senha:
            print("Sucesso")
            
        else:
            print("Senha incorrenta.")
    
    else:
        print(f"Usuário: {login} não existe.")
        


```
