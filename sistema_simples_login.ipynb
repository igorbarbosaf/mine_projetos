```python
# Dicionário vazio
usuarios = {}

# Função menu de escolhas
def menu():

    print("\n1 - Cadastrar usuário")
    print("2 - Fazer login")
    print("3 - Lista de usuários")
    print("4 - Sair")

def cadastrar_usuario():

    login = input("Digite o login: ").strip()

    # Condição para que o login não fique vazio
    if not login:
        print("Login não pode ficar vazio.")
        return

    # Condição para verificar se já tem um login com o mesmo nome
    if login in usuarios:
        print("Login já existe.")
        return
    
    senha = input("Digite sua senha: ")

    # Condição para que senha não fique vazia
    if not senha:
        print("Senha não pode ficar vazia.")
        return

     # Adicionando chave + valor no dicionário   
    usuarios[login] = senha
    print("Usuário cadastrando com sucesso.")

def fazer_login():

    login = input("Digite seu login: ").strip()
    senha = input("Digite sua senha: ")

    # Verificando se o login existe dentro de usuários
    if login in usuarios:

        # Se a senha do usuário for igual a a senha de login então 'sucesso'
        if usuarios[login] == senha:
            print("Sucesso")
            
        else:
            print("Senha incorrenta.")
    
    else:
        print(f"Usuário: {login} não existe.")

def lista_de_usuarios():

    # Verificando se tem algum usuário cadastrando
    if not usuarios:
        print("Nenhum usuário cadastrando")
        return

    # Para usuário e senha usei .item() para buscar chave e valor de dentro do dict 'usuários'
    for usuario, senha in usuarios.items():
            print(f"Usuário: {usuario} - Senha: {senha}")        

while True:
    menu()

    opcao = input("Escolha umas das opções: ")

    if opcao == "1":
        cadastrar_usuario()

    elif opcao == "2":
        fazer_login()

    elif opcao == "3":
        lista_de_usuarios()

    elif opcao == "4":
        print("Fechando...")
        break

    else:
        print("Opção inválida!")

```
