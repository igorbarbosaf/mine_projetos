```python
usuarios = {}

def menu():

    print("\n1 - Cadastrar usuário")
    print("2 - Fazer login")
    print("3 - Visualizar usuário")
    print("4 - Sair")

def cadastrar_usuario():

    login = input("Digite seu login: ").strip()

    # Para evitar de login ficar vazio
    if not login:
        print("Login não pode estar vazio.")
        return

    # Verificando se existe um login igual dentro do dicionário
    if login in usuarios:
        print("Esse login já existe!")
        return

    senha = input("Digite sua senha: ")

    # Validação obrigatória: o campo senha não pode ser nulo
    if not senha:
        print("Senha não pode ficar vazia.")
        return

    # Utilizado len() para percorrer quando caracteres o usuário digitou
    #para que no minimo possa ter apenas 4 caracteres de senha.
    if len(senha) < 4:
        print("Senha deve ter no mínimo 4 caracteres.")
        return

    tentativas = 0 # Contador de erros
    limite = 3 # Limite de tentativas

    # Enquanto tentativas for menor que 3, continue 
    while tentativas < limite:
        
        # Pedido ao usuário confirmar sua senha
        confirmar_senha = input("Confirme a senha: ")

        # Se senha for igual ao digitando então encerra o loop
        if senha == confirmar_senha:
            break

        else:

            # Caso erra a senha tentativas soma +1 no caso tentativas vai valer 1 agora
            tentativas += 1
            print("Você errou a senha! Tente novamente.")

        # Se tentativas chegar ao limite que e 3 então o cadastro e cancelado
        if tentativas == limite:
            print("Você chegou ao limite de tentativas.")
            return

    # Adicionando login'key' e senha'values' dentro do dicionário 'usuarios
    usuarios[login] = senha
    print("Login e senha cadastrando com sucesso!")

def fazer_login():

    login = input("Digite seu login: ").strip()

    # Verificando se o login existe dentro do dict 'usuarios'
    if login not in usuarios:
        print("Esse login não existe.")
        return

    tentativas = 0 # contador vale 0
    limite = 3 # limite vai até 3

    # Enquanto tentativas for menor que o limite continue
    while tentativas < limite:

        senha = input("Digite sua senha: ")

        # Verificando se senha e igual ao values e igual ao que está no dict
        if usuarios[login] == senha:
            print("Login realizando com sucesso.")
            return
        else:
            # Caso erra a senha tentativas vale 1 'Limite de tentativas atigindo.'
            tentativas += 1
            print("Senha incorrenta! Tente novamente.")

    # Se chegar ao limite de tentativas imprime 
    print("Limite de tentativas atigindo.")

def visualizar_usuario():

    # Para login e senha dentro de usuarios função items() para buscar keys e values
    #dentro do dict 'usuarios' e fazer a comparação
    for login, senha in usuarios.items():

        print(f"Login: {login}, Senha: {senha}")

while True:
    menu()

    opcao = input("Digite umas das opção: (1 ao 4): ")

    if opcao == "1":
        cadastrar_usuario()

    elif opcao == "2":
        fazer_login()

    elif opcao == "3":
        visualizar_usuario()

    elif opcao == "4":
        print("Fechando...")
        break
    else:
        print("Opção inválida, porfavor escolha entre 1 ao 4.")
```
