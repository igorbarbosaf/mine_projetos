```python
usuarios = {}

def menu():

    print("\n1 - Cadastrar usuário")
    print("2 - Fazer login")
    print("3 - Visualizar usuário")
    print("4 - Sair")

def cadastrar_usuario():

    login = input("Digite seu login: ").strip()

    # Para evitar de login ficar vazio
    if not login:
        print("Login não pode estar vazio.")
        return

    # Verificando se existe um login igual dentro do dicionário
    if login in usuarios:
        print("Esse login já existe!")
        return

    senha = input("Digite sua senha: ")

    # Validação obrigatória: o campo senha não pode ser nulo
    if not senha:
        print("Senha não pode ficar vazia.")
        return

    # Utilizado len() para percorrer quando caracteres o usuário digitou
    #para que no minimo possa ter apenas 4 caracteres de senha.
    if len(senha) < 4:
        print("Senha deve ter no mínimo 4 caracteres.")
        return

    # Verificando se senha já existe dentro do dicionário
    if senha in usuarios:
        print("Essa senha já existe!")

    else:
        # Adicionando login'key' e senha'values' dentro do dicionário 'usuarios
        usuarios[login] = senha
        print("Login e senha cadastrando com sucesso!")

def fazer_login():

    login = input("Digite seu login: ").strip()
    senha = input("Digite sua senha: ")

    # verificando se o login existe dentro do dicionário 'usuarios
    if login not in usuarios:
        print("Login não existe")

    else:
        # Se login estiver certo ele vai comparar a senha se for igual então entra
        if usuarios[login] == senha:
            print("Login realizando com sucesso")

        else:
            print("Senha incorreta")


def visualizar_usuario():

    # Para login e senha dentro de usuarios função items() para buscar keys e values
    #dentro do dict 'usuarios' e fazer a comparação
    for login, senha in usuarios.items():

        print(f"Login: {login}, Senha: {senha}")

while True:
    menu()

    opcao = input("Digite umas das opção: (1 ao 4): ")

    if opcao == "1":
        cadastrar_usuario()

    elif opcao == "2":
        fazer_login()

    elif opcao == "3":
        visualizar_usuario()

    elif opcao == "4":
        print("Fechando...")
        break
    else:
        print("Opção inválida, porfavor escolha entre 1 ao 4.")
```
