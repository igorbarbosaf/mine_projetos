```python
# Criando uma variável com a função set() vazia
nomes = set()

def menu():

    print("\nMENU")
    print("1 - Cadastrar nome")
    print("2 - Ver nomes")
    print("3 - Sair")


def cadastrar_nome():

    # Pedindo para o usuário digitar nome e usando 'strip' para remover espaço e 'lower' para converte em minúsculo
    nome = input("Digite um nome: ").strip().lower()

    # Caso o usúario deixe o campo vazio
    if not nome:
        print("Porfavor! Esse campo não pode ficar vazio.")
        return

    # Verificando se já existe esse nome dentro do set()
    if nome in nomes:
        print("Nome já existe.")

    else:
        # Adicionado o nome que o usuário digitou dentro do set()
        nomes.add(nome)
        print("Nome Cadastrado")

def ver_nomes():
    # Verificando se tem algum nome cadastrado
    if not nomes:
        print("Nenhum nome cadastrado.")
        return

    # Usando loop 'for' para pecorrer a lista de nome
    for nome in sorted(nomes): # método sorted pra organizar
        print(nome)

# Enquanto for True
while True:
    # Puxando a função menu
    menu()

    opcao = input("Digite um número do menu: ")

    if opcao == "1":
        cadastrar_nome()

    elif opcao == "2":
        ver_nomes()

    elif opcao == "3":
        break

    else:
        print("Por favor! Só digite os números do menu.")
```
