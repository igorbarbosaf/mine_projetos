```python
import random

def jogar():

    numero_secreto = random.randint(1, 50)

    tentetivas = 0 # Contador começa
    limite = 5 # Limite vai até 5

    # Enquanto tentativas for menor que limite
    while tentetivas < limite:

        try:

            secreto = int(input("Digite um número: "))

        # Verificando se o usuário não digitou nada além de números inteiros, caso sim dar Erro
        except ValueError:
            print("Por favor! Digite apenas números.")
            continue
        
        if secreto == numero_secreto:
            print("Parabéns! Você acertou o número", numero_secreto)
            break
        
        else:
            tentetivas += 1

            if secreto > numero_secreto:
                print("O número é menor.")
            else:
                print("O número é maior.") 

jogar()

        
```
